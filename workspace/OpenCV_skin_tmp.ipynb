```python
test1 = './like_lenna.png'
```

```python
img = cv2.imread(test1)

# 이미지 색상 변환 (OpenCV는 BGR 형식이므로 RGB로 변환 필요)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지 출력
plt.imshow(img_rgb)
plt.axis('off')  # 축 숨기기
plt.show()
```

```python
# 이미지 불러오기 (그레이스케일로 변환)
img = cv2.imread(test1, cv2.IMREAD_GRAYSCALE)

# 캐니 엣지 검출 => **가장 많이 쓰이는 엣지 검출 방법
# (이미지, threshold1, threshold2)
# threshold1 -> 값이 작다면 : 약한 선도 표현
# threshold2 -> 값이 크다면 : 뚜렷한 선만 표현
edges = cv2.Canny(img, 50, 250)

# 원본 이미지와 엣지 검출된 이미지 출력
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

```python
# 이미지 불러오기
img = cv2.imread(test1)

# 그레이스케일로 변환
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 캐니 엣지 검출로 외곽선 강조
edges = cv2.Canny(gray, 100, 200)

# 외곽선 찾기 (윤곽선을 찾고 계층 구조 정보는 사용하지 않음) => 대상의 경계를 파악하는 데 자주 씀
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 원본 이미지에 외곽선 그리기
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)  # 초록색 외곽선, 두께 2

# 결과 출력
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.title('Contours')
plt.axis('off')

plt.show()
```

```python
# 이미지 불러오기
img = cv2.imread(test3)

# 그레이스케일로 변환
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 캐니 엣지 검출로 외곽선 강조
edges = cv2.Canny(gray, 100, 200)

# 외곽선 찾기 (윤곽선을 찾고 계층 구조 정보는 사용하지 않음)
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 원본 이미지에 외곽선 그리기
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)  # 초록색 외곽선, 두께 2

# 결과 출력
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.title('Contours')
plt.axis('off')

plt.show()
```

```python
# 이미지 불러오기
img = cv2.imread(test1)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# 빨간색 범위 설정 (낮은 빨간색과 높은 빨간색 두 범위 사용)
lower_red1 = np.array([0, 100, 100])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([160, 100, 100])
upper_red2 = np.array([180, 255, 255])

# 빨간색 영역 마스크 생성
mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
red_mask = cv2.bitwise_or(mask1, mask2)

# 원본 이미지에서 빨간색 영역만 추출
red_only = cv2.bitwise_and(img, img, mask=red_mask)

# 결과 출력
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(red_mask, cmap='gray')
plt.title('Red Mask')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(red_only, cv2.COLOR_BGR2RGB))
plt.title('Red Color Detection')
plt.axis('off')

plt.show()
```

```python
# 이미지 불러오기
img = cv2.imread(test2)

# 이미지를 BGR에서 HSV 색 공간으로 변환
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# 검은색과 갈색 범위 설정 (HSV 값은 조명에 따라 조정 가능)
# 검은색 범위
lower_black = np.array([0, 0, 0])
upper_black = np.array([180, 255, 50])

# 갈색 범위
lower_brown = np.array([10, 100, 20])
upper_brown = np.array([20, 255, 200])

# 검은색 및 갈색 마스크 생성
black_mask = cv2.inRange(hsv, lower_black, upper_black)
brown_mask = cv2.inRange(hsv, lower_brown, upper_brown)

# 검은색과 갈색 마스크 결합
mask = cv2.bitwise_or(black_mask, brown_mask)

# 원본 이미지에서 검은색-갈색 영역만 추출
mole_only = cv2.bitwise_and(img, img, mask=mask)

# 결과 출력
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title('Black-Brown Mask')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(mole_only, cv2.COLOR_BGR2RGB))
plt.title('Mole Detection')
plt.axis('off')

plt.show()
```
